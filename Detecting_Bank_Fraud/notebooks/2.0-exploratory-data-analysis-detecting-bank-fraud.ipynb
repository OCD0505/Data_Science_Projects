#### 2.5.1.1 Exploration of Transaction Amount and Fraud Relation <a id='2.5.1.1_Transaction_Amount_and_Fraud_Relation'><a/>

```python
# To begin, thresholds must be defined to determine what is small vs large

# Quick look at the stats for the amount column
BankData['amount'].describe()
```

The dataset has 6.36 million entries, with a mean transaction amount of 179,861.9 and a standard deviation of 603,858.2. The minimum value is 0, and the maximum is 92.45 million. The 25th percentile is 13,389.6, the 50th percentile (median) is 74,871.9, and the 75th percentile is 208,721.5. This suggests a highly skewed distribution with a few very large transactions.








```python
# Filter data to exclude extreme outliers (e.g., top 1%)
# outliers removed to generate clear visualization
threshold = BankData['amount'].quantile(0.99)
filtered_data = BankData[BankData['amount'] <= threshold]

# Histogram of filtered data
plt.figure(figsize=(12, 6))
sns.histplot(filtered_data['amount'], bins=50, kde=True, color='skyblue')
plt.title("Histogram of Amounts (Filtered Below 99th Percentile)")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.axvline(filtered_data['amount'].quantile(0.25), color='red', linestyle='--', label='25th Percentile')
plt.axvline(filtered_data['amount'].median(), color='green', linestyle='--', label='Median (50th Percentile)')
plt.axvline(filtered_data['amount'].quantile(0.75), color='blue', linestyle='--', label='75th Percentile')
plt.legend()
plt.show()
```

```python
# Define thresholds
q25 = filtered_data['amount'].quantile(0.25)
q75 = filtered_data['amount'].quantile(0.75)

# Ensure we're working on a copy explicitly
filtered_data = filtered_data.copy()

# Create the 'size_category' column using pd.cut
filtered_data['size_category'] = pd.cut(
    filtered_data['amount'],
    bins=[-float('inf'), q25, q75, float('inf')],  # Define bins
    labels=['Small', 'Medium', 'Large']            # Define labels
)

# Display thresholds
print(f"Small: Below {q25}")
print(f"Medium: Between {q25} and {q75}")
print(f"Large: Above {q75}")
```

The data has been categorized into three size groups based on transaction amounts:

Small: Transactions with amounts below 13,189.63 are classified as small.
Medium: Transactions ranging from 13,189.63 to 203,647.35 fall into the medium category.
Large: Transactions exceeding 203,647.35 are classified as large.

Using these ranges, correlation will be explored to see if there is correlation between amount size and fraud.

```python
# Filter data by size categories
small_data = filtered_data[filtered_data['size_category'] == 'Small']
medium_data = filtered_data[filtered_data['size_category'] == 'Medium']
large_data = filtered_data[filtered_data['size_category'] == 'Large']

# Compute correlations
small_corr = small_data[['amount', 'isFraud']].corr().loc['amount', 'isFraud']
medium_corr = medium_data[['amount', 'isFraud']].corr().loc['amount', 'isFraud']
large_corr = large_data[['amount', 'isFraud']].corr().loc['amount', 'isFraud']

# Output the results
print("Correlation between 'amount' and 'isFraud':")
print(f"Small Transactions: {small_corr}")
print(f"Medium Transactions: {medium_corr}")
print(f"Large Transactions: {large_corr}")
```

The correlations for small and medium transactions are close to zero, indicating little to no relationship between transaction amount and fraud. For large transactions, there is a weak positive correlation (0.0533), suggesting a slight increase in fraud likelihood with higher amounts, but the relationship remains weak and not highly significant.

#### 2.5.1.2 Outlier Analysis and Handling <a id='2.5.1.2_Outlier_Analysis_and_Handling'><a/>

In the previous step, a threshold of 1,615,979.47 was defined using the 99th percentile of the dataset (i.e., BankData['amount'].quantile(0.99)) to identify outliers. Amounts greater than this value are considered outliers. In this section, the relationship between these outliers (amounts exceeding the threshold) and the isFraud column will be explored. Although size categories were created in the previous step, outliers were excluded from this analysis.

```python
# Data setup
threshold = BankData['amount'].quantile(0.99)
filtered_data = BankData[BankData['amount'] <= threshold]
outliers = BankData[BankData['amount'] > threshold]

# Log scale visualization
plt.figure(figsize=(12, 6))

# Plot the histogram with log scale
histplot=sns.histplot(filtered_data['amount'], bins=30, kde=True, color='blue', label='Filtered Data')

# Add the threshold line
plt.axvline(threshold, color='red', linestyle='--', \
            label=f'Threshold (99th Percentile) = {threshold:.2f}')

# Annotate the threshold
plt.text(threshold, 1, '99th Percentile', color='red', fontsize=12, rotation=90)

# Overlay outliers
plt.scatter(outliers['amount'], [0] * len(outliers), color='orange', \
            label='Outliers', marker='x')

# Apply log scale to x-axis
plt.xscale('log')

# Add titles and legend
plt.title('Distribution of Amounts (Log Scale) with Outliers Highlighted', fontsize=14)
plt.xlabel('Amount (Log Scale)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.tight_layout()

plt.show()
```

- The red vertical line marks the 99th percentile, which is a threshold used to identify outliers. Any data points to the right of this line are considered outliers, as they are significantly larger than the majority of the data.

- The blue bars represent the distribution of the filtered data (without the outliers). The shape of the distribution is skewed to the right, indicating that most of the values are relatively small, with a few larger values.
- The log scale on the x-axis compresses the larger values, making it easier to visualize the overall distribution.

```python
# Calculate the correlation between 'amount' and 'isFraud' for the filtered data
correlation = outliers['amount'].corr(outliers['isFraud'])

# Print the correlation value
print(correlation)
```

The correlation between the outliers (transactions with amounts above the 99th percentile threshold) and the fraud indicator (isFraud) is 0.0379, which is very close to zero.

This indicates a very weak positive correlation between the size of the outlier transactions and the likelihood of fraud. In other words, the size of the transaction (as identified by the 99th percentile threshold) does not significantly influence whether a transaction is flagged as fraudulent. This aligns with the findings from the previous analysis, which also found no strong relationship between fraud and transaction size, even when outliers were excluded.

#### 2.5.1.3 Class Imbalance Impact <a id='2.5.1.3_Class_Imbalance_Impact'><a/>

To analyze class imbalance in the dataset, with respect to the isFraud column, how balanced or imbalanced the distribution of values is in the target variable (isFraud) will need to be assessed. In the case of fraud detection, this is typically a binary classification problem (fraud = 1, non-fraud = 0), and class imbalance often arises when fraudulent transactions are much fewer than non-fraudulent ones.

```python
# 1. Distribution of the target variable 'isFraud'
fraud_distribution = BankData['isFraud'].value_counts()

# 2. Plot the distribution of classes (fraud vs non-fraud)
plt.figure(figsize=(8, 6))
sns.countplot(x='isFraud', data=BankData, palette='Set2')
plt.title('Class Distribution: Fraud vs Non-Fraud', fontsize=14)
plt.xlabel('Class (0 = Non-Fraud, 1 = Fraud)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 3. Pie chart for visual representation
plt.figure(figsize=(8, 6))
fraud_distribution.plot.pie(autopct='%1.1f%%', colors=['lightblue', 'salmon'], labels=['Non-Fraud', 'Fraud'], startangle=90, legend=True)
plt.title('Class Distribution: Fraud vs Non-Fraud (Pie Chart)', fontsize=14)
plt.ylabel('')
plt.show()

# 4. Calculate the imbalance ratio (fraud:non-fraud)
fraud_count = fraud_distribution[1]  # Count of fraud (1)
non_fraud_count = fraud_distribution[0]  # Count of non-fraud (0)

imbalance_ratio = fraud_count / non_fraud_count if non_fraud_count != 0 else float('inf')
print(f"Imbalance ratio (Fraud:Non-Fraud): {imbalance_ratio:.3f}")
```

The dataset shows a significant class imbalance, with 99.9% of transactions being non-fraudulent and only 0.1% fraudulent. This imbalance makes it challenging for a model to detect fraud, as it may be biased towards predicting the majority class (non-fraudulent). This can result in low sensitivity and high false negatives.

To address this, several techniques can be applied. Data augmentation (such as oversampling fraudulent transactions or undersampling non-fraudulent ones) can help balance the dataset. Class weighting assigns higher importance to the fraudulent class during training, while ensemble methods combine multiple models for improved performance. SMOTE or ADASYN can generate synthetic data for the minority class to enhance learning.

The next steps include ensuring data quality, experimenting with resampling and class weighting, and evaluating the model with metrics like precision, recall, and AUC-ROC. By applying these methods, the model's ability to detect fraud will improve despite the imbalance.

```python
#count of fraudulent transactions per transaction type

ax = BankData.groupby('type').agg({"isFraud":"sum"}).plot(kind='bar', rot=45)
plt.title("Count of fraudulent transactions \nper transaction type\n")

# Get current Y-axis tick positions and labels
yticks = ax.get_yticks()

# Update Y-axis tick labels to show values in millions
new_labels = [f'{float(y / 1000000)}M' for y in yticks]
_=ax.set_yticks(yticks)
_=ax.set_yticklabels(new_labels)
```

The CASH_OUT and PAYMENT transaction types exhibit all the fraudulent activity.

```python
# Set the figure size
plt.figure(figsize=(14, 6))

# Create box plot with log-transformed 'amount', using 'hue' to separate fraudulent and non-fraudulent transactions
sns.boxplot(x='type', y='log_amount', hue='isFraud', data=BankData, 
            palette={0: "blue", 1: "red"}, hue_order=[0, 1])

# Title and labels
plt.title("Distribution of Log-Transformed 'Amount' by Transaction Type (Fraudulent vs Non-Fraudulent)")
plt.xlabel("Transaction Type")
plt.ylabel("Log(Transaction Amount)")
plt.xticks(rotation=45)

# ensure the color matches in legened
handles, labels = plt.gca().get_legend_handles_labels()  # Get current legend handles and labels
plt.legend(handles=handles, labels=["Non-Fraudulent", "Fraudulent"], title="Fraudulent")

plt.show()
```

Fraudulent TRANSFER transactions (red) generally have slightly higher median amounts than non-fraudulent ones (blue).There is a wider spread for both transaction types, with significant outliers in both groups.

Fraudulent CASH_OUT transactions have a higher median log amount compared to non-fraudulent ones, indicating they tend to involve larger sums. Similar to TRANSFER, there is a broader range and a high number of outliers for both categories.

**Reminder**: This is an artificial dataset where all fraudulent transactions were intentionally designed by the creator. This could reflect a deliberate design choice to mimic real-world scenarios where these types are more prone to fraud due to their inherent risks (e.g., transferring funds to external accounts or cashing out large sums).

The artificial dataset design, where fraudulent transactions are exclusively tied to specific transaction types (TRANSFER and CASH_OUT), will significantly influence the modeling stage. Key implications include:

- Transaction type will emerge as a highly predictive feature for fraud detection.
This may cause the model to overfit, relying too heavily on transaction type and ignoring other important fraud indicators.

- The model might struggle to generalize to real-world scenarios where fraud can occur across other transaction types (PAYMENT, DEBIT, CASH_IN), leading to biased or incomplete predictions.

- The lack of fraudulent examples in non-TRANSFER/CASH_OUT types limits the model’s ability to detect fraud in diverse contexts, potentially reducing its effectiveness on real-world data.

Potential Solutions:

- Introduce fraudulent examples in other transaction types to simulate real-world diversity.
Feature engineering: Incorporate additional features (e.g., time of transaction, user behavior, location) to reduce reliance on transaction type.
- Use techniques to minimize overfitting and ensure the model considers multiple features when predicting fraud.

## 2.6 Summary <a id='2.6_Summary'></a>


The analysis of this artificial dataset, designed to simulate fraud detection in banking transactions, provides key insights into the structure of the data and its potential implications for modeling. The findings highlight several patterns and challenges that must be addressed before proceeding to the preprocessing phase.

- The dataset includes five transaction types: CASH-IN, CASH-OUT, DEBIT, PAYMENT, TRANSFER.
Fraudulent transactions were deliberately limited to TRANSFER and CASH_OUT types, reflecting real-world scenarios where these types are more prone to fraudulent behavior.

- Fraud is simulated by agents aiming to empty accounts through a sequence of TRANSFER followed by CASH_OUT transactions. Fraudulent transactions tend to have higher median transaction amounts and broader distributions, with many outliers.

- The transaction amounts range widely, with a mean of 179,861.9 and a highly skewed distribution. Transactions have been categorized into small, medium, and large sizes, but no significant correlation was found between transaction size and fraud likelihood, even for outliers.

- The dataset is highly imbalanced, with only 0.1% of transactions flagged as fraudulent. This poses challenges for modeling, as models may bias toward the majority non-fraudulent class.

**Challenges**

- Fraudulent activity is tied exclusively to TRANSFER and CASH_OUT, making transaction type a dominant predictive feature. This could lead to overfitting, where the model relies too heavily on transaction type and ignores other potentially important factors.

- While outliers (amounts exceeding the 99th percentile) show a weak correlation with fraud (0.0379), they still represent a unique subset of data that may require careful handling.
Limited Real-World Representation:

- The artificial design may not reflect the complexity of real-world fraud patterns, where fraud can occur across all transaction types and involve subtler indicators.


**Next Steps**

To address these challenges and prepare the dataset for robust modeling, the following preprocessing steps are planned:

- Use techniques such as oversampling (e.g., SMOTE or ADASYN) or undersampling to address class imbalance and ensure the model learns effectively from the minority class.


- Introduce additional features (e.g., transaction time, user behavior, historical patterns) to reduce reliance on transaction type and enhance predictive performance.


- Evaluate the impact of outliers on the model by normalizing or capping extreme values where necessary to minimize their influence on learning.


- Apply scaling (e.g., MinMax or StandardScaler) to ensure numerical features like transaction amounts are appropriately normalized for machine learning algorithms.


- Implement methods like class weighting or ensemble learning to mitigate overfitting and encourage the model to learn broader patterns beyond transaction type.
Validation Strategy:

- Use stratified sampling or k-fold cross-validation to evaluate model performance, focusing on metrics such as precision, recall, and AUC-ROC to account for the imbalanced dataset.
